```python
# Load required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import datetime as dt
import seaborn as sns

# Show plots in the notebook.
%matplotlib notebook 

# Read New Property date into a data frame
df1 = pd.read_csv(r'./new.csv')
df1 = df1[['Month', 'New Properties']]

# Read Existing Property date into a data frame
df2 = pd.read_csv(r'./existing.csv')
df2 = df2[['Month', 'Existing Properties']]

# Plot the results using Seaborn to use built-in trend lines
sns.regplot(df1['Month'],df1['New Properties'], label= "New Properties")
sns.regplot(df2['Month'],df2['Existing Properties'], label= "Existing Properties")
plt.ylabel('Loan Amount (in Millions)', fontsize=12)
plt.xlabel('Months since 1982', fontsize=12)
plt.title("Canadian Mortgage Loan Values")
plt.legend(loc='upper left')
```
